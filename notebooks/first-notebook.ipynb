```python
wc_teams = list(world_cup_squads['Team'].value_counts().index)
teams_avg_age = []
teams_max_age = []
teams_min_age = []
for team in wc_teams:
    x = world_cup_squads[world_cup_squads['Team']==team][['Age']].sum()/26
    teams_avg_age.append(int(x))
    y = world_cup_squads[world_cup_squads['Team']==team][['Age']].max()
    teams_max_age.append(int(y))
    z = world_cup_squads[world_cup_squads['Team']==team][['Age']].min()
    teams_min_age.append(int(z))
        
```

```python
new_ii = {'Team':wc_teams,
          'Min Age':teams_min_age,
          'Max Age':teams_max_age,
          'Avg Age':teams_avg_age
         }
_age = pd.DataFrame(new_ii)
```
